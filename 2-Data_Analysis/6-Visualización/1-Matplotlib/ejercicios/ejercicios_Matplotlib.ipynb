2. Sabemos que podemos pintar gráficas de dos formas: la figura incluye los ejes o teniendo figura y ejes por separado.

Usando solo una figura, usa numpy para los valores del eje X entre 0 y 5. 

Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). 

Elige la precisión en el eje X que desees.  

```python
x = np.linspace(0,5,100)

fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,4))
ax1.plot(x, 3 * x, color='blue')
ax1.set_title('Recta: y = 3x')
ax2.plot(x, -3 * x - 5, color='red')
ax2.set_title('Recta: y = -3x - 5')
plt.tight_layout()
plt.show()
```

5. Usando una figura Y EJES POR SEPARADO, usa numpy para los valores del eje X entre 0 y 5. Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). Elige la precisión en el eje X que desees.  

```python
fig = plt.figure(figsize=(10, 4))


x = np.linspace(0, 5, 100)

# --- Primer eje (izquierda): Recta y = 3x ---
ax1 = fig.add_subplot(121) 
ax1.plot(x, 3 * x, color='blue')
ax1.set_title('Recta: y = 3x')
ax1.set_xlabel('eje X')
ax1.set_ylabel('eje Y')

ax1.grid(True)

# --- Segundo eje (derecha): Recta y = -3x - 5 ---
ax2 = fig.add_subplot(122) 
ax2.plot(x, -3 * x - 5, color='red')
ax2.set_title('Recta: y = -3x - 5')
ax2.set_xlabel('eje X')

ax2.grid(True)

# Ajustar espaciado y mostrar
plt.tight_layout()
plt.show()
```

6. Crea una lista de coordenadas X: 20, 22, 24, 26, 28
Crea una lista de coordenadas Y: 5, 15, -5, 20, 5
Usa un gráfico de barras colocadas en X con alturas Y

```python
coordenadas_x = [20, 22, 24, 26, 28]
coordenadas_y = [5, 15, -5, 20, 5]

plt.figure(figsize=(8, 5)) 


barras = plt.bar(coordenadas_x, coordenadas_y, 
                 width=1.5, 
                 color='skyblue', 
                 edgecolor='navy',  
                 linewidth=1)  

# Personalización adicional
plt.title('Gráfico de Barras en alturas', pad=20)
plt.xlabel('Coordenadas X', labelpad=10)
plt.ylabel('Coordenadas Y', labelpad=10)
```

13. Vamos a introducir pandas. Crea un DataFrame con las columnas "Year" de valores 2015, 2016, 2017, 2018, 2019 y la columna "Sold_items_A" de valores 1000, 3500, 4000, 5500, 7000

```python
import pandas as pd 

data = {
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Sold_items_A': [1000, 3500, 4000, 5500, 7000]
}

df = pd.DataFrame(data)
```

14. Ahora pintamos un gráfico de línea con las ventas respecto al año. Pon un título y etiquetas en los ejes.
Cambia el estilo a 'seaborn-white'. La línea debe ser con rayas y verde.

```python
fig, ax = plt.subplots(figsize=(8, 5))

import seaborn as sns
sns.set_style("white")


ax.plot(data['Year'], data['Sold_items_A'], 
        color='green', 
        linestyle='--',  
        linewidth=2, 
        marker='o',  
        markersize=8,
        markerfacecolor='white',  
        markeredgecolor='green') 

# Personalización adicional
plt.title('Ventas por año', pad=20)
plt.xlabel('Año', labelpad=10)
plt.ylabel('Ventas', labelpad=10)
```

15. Otro departamento B ha vendido en esos años 2000, 3100, 5000, 4000, 6000 unidades. Incluye esa columna en el DataFrame y pinta en la misma gráfica las dos líneas. B es una línea punteada y roja. Muestra una leyenda abajo a la derecha.

```python
data = {
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Sold_items_A': [1000, 3500, 4000, 5500, 7000],
    'Sold_items_B': [2000, 3100, 5000, 4000, 6000]
}

# 2. Configuración del gráfico
sns.set_style("white")
fig, ax = plt.subplots(figsize=(8, 5))

# 3. Gráfico para Producto A (verde)
ax.plot(data['Year'], data['Sold_items_A'], 
        color='green',
        linestyle='--',  # Línea discontinua
        linewidth=2,
        marker='o',      # Marcadores circulares
        markersize=8,
        markerfacecolor='white',
        markeredgecolor='green',
        label='Departamento A')

# 4. Gráfico para Producto B (rojo) - ¡Línea separada!
ax.plot(data['Year'], data['Sold_items_B'],
        color='red',
        linestyle=':',   # Línea punteada
        linewidth=2,
        marker='o',
        markersize=8,
        markerfacecolor='white',
        markeredgecolor='red',
        label='Departamento B')

# 5. Personalización
ax.set_title('Ventas anuales por producto', pad=20, fontweight='bold')
ax.set_xlabel('Año')
ax.set_ylabel('Unidades vendidas')
ax.legend(loc='lower right')
ax.grid(True, linestyle=':', alpha=0.4)

plt.tight_layout()
plt.show()
```

19. Vamos a pintar un histograma de unas alturas generadas aleatoriamente. Crea 300 alturas entre 150 y 190 cm.

```python
import random

alturas = np.random.uniform(150,190,300).round(2).T
plt.hist(alturas, 
         bins=15,  # Número de barras
         color='skyblue', 
         edgecolor='navy',
         linewidth=1.2,
         alpha=0.8)

# 3. Personalización avanzada
plt.title('Distribución de Alturas (n=300)', fontsize=14, pad=20)
plt.xlabel('Altura (cm)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.xticks(np.arange(150, 191, 5)) 
```
