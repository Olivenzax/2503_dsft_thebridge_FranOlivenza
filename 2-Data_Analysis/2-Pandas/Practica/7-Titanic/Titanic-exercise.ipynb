9. What is the name of the oldest person who died in the Titanic? Was he or she travelling alone or had any family travelling with them?  



```python
died_passengers = titanic[titanic['Survived'] == 0]

oldest_person = died_passengers.loc[died_passengers['Age'].idxmax()]

name = oldest_person['Name']
age = oldest_person['Age']
sibsp = oldest_person['SibSp']  # Number of siblings/spouses aboard
parch = oldest_person['Parch']  # Number of parents/children aboard

# Determine if they were traveling alone or with family
if sibsp == 0 and parch == 0:
    travel_status = "Viajaba solo"
else:
    travel_status = f"Viajaba con su familia."

# Print the result
print(f"La persona m√°s vieja que muri√≥ fue {name}, a la edad de {age}. {travel_status}.")
```

21.  Group `titanic` by `'pclass'` and aggregate by the columns `age` & `fare`, by `max` and `median` and assign it to `by_class`


```python
by_class = titanic.groupby('Pclass').agg(
    max_age=('Age', 'max'),
    median_age=('Age', 'median'),
    max_fare=('Fare', 'max'),
    median_fare=('Fare', 'median')
)

# Reset index if you want `pclass` as a column instead of the index
by_class = by_class.reset_index()
by_class
```

24. Using [`.pivot_table()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html) to count how many women or men survived by class, call it `counted`.

Don't panic and read the documentation!




```python
import pandas as pd

counted = pd.pivot_table(
    titanic,
    values='Survived',  
    index='Pclass',     
    columns='gender',      
    aggfunc='sum'       
)

counted = counted.reset_index()

counted
```

25. Add a new column with the sum of survived men and women, call it `counted['total']` 


```python
counted['total'] = counted['female'] + counted['male']

counted
```

26. Sort `counted` by the `'total'` column. In which class the people survived the most?
