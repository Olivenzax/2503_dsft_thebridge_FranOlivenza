## Ejercicio 18

## Ejercicio 20
Agrupa por estado y ciudad, la media, mediana y máximo nivel de alcohol

```python
ciudades_estados = cervezas_completo.groupby(['state', 'city'])['abv'].agg(
    media_abv='mean',     # Media de alcohol
    mediana_abv='median', # Mediana
    max_abv='max'         # Máximo nivel de alcohol
).reset_index()

ciudades_estados = ciudades_estados.sort_values(by='max_abv', ascending=False)

# Mostrar los 10 primeros resultados
print(ciudades_estados.head(10))
```

## Ejercicio 23
Consigue en un DataFrame de todas las cervezas en cuyas cervecerías al menos haya una cerveza con un `abv` mínimo 0.07.

```python
# todas las cervezas, que venda una cervecería, donde al menos una de ellas sea de 0.07 para arriba
```

```python
cervezas_mas_alcohol = cervezas_completo[cervezas_completo['brewery_name'].isin(cervezas_completo[cervezas_completo['abv']>=0.07]['brewery_name'].unique())][['name','brewery_name','abv']]

cervezas_mas_alcohol
```

## Ejercicio 24
Teniendo en cuenta la media de amargor de todas las cervezas, obtén una tabla con las cervezas cuyos estados tengan una media de amargor en cerveza por encima de la media

```python
# Las cervezas que se encuentran en aquellos estados, cuya media de amargor, sea superior a la media de amargor total
```

```python
media_global_amargor = cervezas_completo['ibu'].mean()

media_estado_amargor = cervezas_completo.groupby('state')['ibu'].mean()

estados_alto_amargor = media_estado_amargor[media_estado_amargor > media_global_amargor].index

cervezas_filtradas = cervezas_completo[cervezas_completo['state'].isin(estados_alto_amargor)]

cervezas_filtradas[['name', 'brewery_name', 'state', 'ibu']]
```
