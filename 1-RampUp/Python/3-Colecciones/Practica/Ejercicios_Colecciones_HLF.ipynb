## Ejercicio Extra HLF
Vamos a empezar a programar trozos de código que nos servirán para el HLF:

* Crear un Tablero como una lista de listas de 10x10 que contenga caracteres espacio: " "
* Insertar dos barcos sustituyendo en el Tablero los espacios en blanco por el caracter "B":
    * Uno de 4 casillas en horizontal: casillas [1][0], [1][1], [1][2], [1][3]
    * Uno de 3 casillas en vertical:   casillas [3][3], [4][3], [4][4], [4][5]
* Escribir una celda de código separada de las anteriores para simular que el usuario está efectuando un disparo: 
    *  Solicite al usuario dos coordenadas `i,j` mediante dos inputs
    *  Compruebe si esas coordenadas están ocupada por una casilla barco (que haya una en "B")
        * Si hay una "B":
            * Cambiar el contenido de esa posición por una "x"
            * Imprimir por pantalla el texto "Tocado en posición `i,j`"
        * Si hay un " ":
            * Cambiar el contenido de esa posición por una "o"
            * Imprimir por pantalla el texto "Agua"
    * Imprimir por pantalla el Tablero
* Para comprobar que funciona bien la celda de disparar ejecutarla varias veces metiendo distintas coordenadas

### Bonus
* En la celda de disparar, si el usuario acierta a tocar un barco, vuelve a disparar
* Restilizando funciones
    


```python
# Ejercicio Extra HLF
print("¡Bienvenido a Hundir la Flota!\n\n\t Introduzca las coordenadas\n\n")

# Crear un tablero como lista de listas de 10x10 que contengan caracteres espacio
tablero = [[" " for _ in range(10)] for _ in range(10)]

# Función de disparar
def disparar(tablero):
    i = int(input("Introduce la coordenada i (fila): "))
    j = int(input("Introduce la coordenada j (columna): "))
    print(f"Has disparado a la posición [{i},{j}] y el resultado es:")
    # Solicitar coordenadas al usuario
    if 0 <= i < 10 and 0 <= j < 10:
        if tablero[i][j] == "B":
            tablero[i][j] = "X" 
            print(f"Tocado en posición {i},{j}")
        elif tablero[i][j] == " ":
            tablero[i][j] = "A" 
            print("Agua")
        else:
            print("Ya has disparado en esta posición.")
    else:
        print("Coordenadas fuera del rango del tablero.")
    print()
    # Imprimir la función de disparo para probar
    for fila in tablero:
        print("|".join(fila))
    print("\n")

# Barcos
#* Insertar dos barcos sustituyendo en el Tablero los espacios en blanco por el caracter "B":
##   * Uno de 4 casillas en horizontal: casillas [1][0], [1][1], [1][2], [1][3]
##   * Uno de 3 casillas en vertical:   casillas [3][3], [4][3], [4][4], [4][5]

for j in range(4):
    tablero[1][j] = "B"

for i in range(3, 6):
    tablero[i][3] = "B"


while any("B" in fila for fila in tablero):
        disparar(tablero)
print("¡Felicidades! Has hundido todos los barcos. ¡Ganaste!")
```

```python
# Ejercicio Extra HLF (Se ven solo las que disparas)
print("¡Bienvenido a Hundir la Flota!\n\n\t Introduzca las coordenadas\n\n")

# Crear dos tableros (uno visible y otro oculto) como lista de listas de 10x10 que contengan caracteres espacio
tablero_oculto = [[" " for _ in range(10)] for _ in range(10)]
tablero_visible = [[" " for _ in range(10)] for _ in range(10)]

# Función de disparar
def disparar(tablero_oculto, tablero_visible):
    i = int(input("Introduce la coordenada i (fila): "))
    j = int(input("Introduce la coordenada j (columna): "))
    print(f"Has disparado a la posición [{i},{j}] y el resultado es:")
    # Solicitar coordenadas al usuario
    if 0 <= i < 10 and 0 <= j < 10:
        if tablero_oculto[i][j] == "B":
            tablero_oculto[i][j] = "X" 
            tablero_visible[i][j] = "X" 
            print(f"Tocado en posición {i},{j}")
        elif tablero_oculto[i][j] == " ":
            tablero_oculto[i][j] = "A"
            tablero_visible[i][j] = "A"  
            print("Agua")
        else:
            print("Ya has disparado en esta posición.")
    else:
        print("Coordenadas fuera del rango del tablero.")
    print()
    # Imprimir la función de disparo para probar
    for fila in tablero_visible:
        print("|".join(fila))
    print("\n")

# Barcos
#* Insertar dos barcos sustituyendo en el Tablero los espacios en blanco por el caracter "B":
##   * Uno de 4 casillas en horizontal: casillas [1][0], [1][1], [1][2], [1][3]
##   * Uno de 3 casillas en vertical:   casillas [3][3], [4][3], [4][4], [4][5]

for j in range(4):
    tablero_oculto[1][j] = "B"

for i in range(3, 6):
    tablero_oculto[i][3] = "B"


while any("B" in fila for fila in tablero_oculto):
        disparar(tablero_oculto, tablero_visible)
print("¡Felicidades! Has hundido todos los barcos. ¡Ganaste!")
```
